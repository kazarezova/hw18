```python
N = 20

x = np.linspace(-5, 5, N)
y = 3*x - 1 + np.random.normal(scale=10, size=N)

X = np.column_stack([
    x**i for i in range(N)
])

model = LinearRegression(fit_intercept=False)
model.fit(X, y)
print(model.score(X, y))

print(model.coef_)
print(model.intercept_)

y_pred = model.predict(X)

sns.scatterplot(x=x, y=y)
sns.lineplot(x=x, y=y_pred)
plt.ylim([-10, 10])
```

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, Lasso, LassoCV

import math
```

```python
df = pd.read_csv("sgn.csv")

x = df['x'].to_numpy()
y = df['y'].to_numpy()

m=1
 
X=[]
for each in x:
    new_row=[]
    for i in range(1,m+1):
        new_row.append(math.sin(each*i))
        new_row.append(math.cos(each*i))
    X.append(new_row)
X=pd.DataFrame(X).to_numpy()

model = LinearRegression()
model.fit(X, y)
print('score:',model.score(X, y))

print('coef:',model.coef_)
print('intercept:',model.intercept_)

y_pred = model.predict(X)

sns.scatterplot(x=x, y=y)
sns.lineplot(x=x, y=y_pred)
```

# 3
–†–µ—à–∏—Ç—å –ø—Ä–æ—à–ª—É—é –∑–∞–¥–∞—á—É –ø—Ä–∏ ùëö=1000, –∏—Å–ø–æ–ª—å–∑—É—è LassoCV –≤–º–µ—Å—Ç–æ LinearRegression. –í–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –∏ –∏–Ω—Ç–µ—Ä–ø—Ä–µ—Ç–∏—Ä–æ–≤–∞—Ç—å —Ä–µ–∑—É–ª—å—Ç–∞—Ç.

```python
df = pd.read_csv("sgn.csv")

x = df['x'].to_numpy()
y = df['y'].to_numpy()

m=1000
 
X=[]
for each in x:
    new_row=[]
    for i in range(1,m+1):
        new_row.append(math.sin(each*i))
        new_row.append(math.cos(each*i))
    X.append(new_row)
X=pd.DataFrame(X).to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=17)

model = LassoCV(positive=True)
model.fit(X_train, y_train)

print('train score:',model.score(X_train, y_train))
print('test score:',model.score(X_test, y_test))

print('coef: too many of them') #model.coef_
print('intercept:',model.intercept_)

y_pred = model.predict(X)

sns.scatterplot(x=x, y=y)
sns.lineplot(x=x, y=y_pred)
```
